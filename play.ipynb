# Projects

### Brief summary — Projects section

* Auto-shows your GitHub repos in “My Projects”
* Default: **9 most recently updated (pushed)** repos
* Forked repos are **excluded by default**

### Config controls (`_config.yml`)

* `sort_by`: how repos are ordered

  * `pushed` (recent activity)
  * `stars` (popularity)
* `limit`: number of repos shown (default 9)
* `exclude`:

  * `forks: true` → hide forked repos
  * `archived: true` → hide archived repos
  * `projects:` → manually hide specific repo names

# Social Media 

**Social media (summary):**

Used to add links to your profiles (LinkedIn, Twitter, GitHub, YouTube, etc.) in your website.

### How it works:

* Set accounts in `_config.yml`
* Format: `platform: username/link`

### Purpose:

* Lets visitors quickly find your profiles
* Displays social icons/links on your site


#Social media links appear in two places:

Header → shows your profile links next to your bio
Blog posts → adds share buttons (if supported)

Purpose:
Connect your site to your social profiles and enable easy sharing of posts.


#Header = the top section of a website




Two files control social media features:

* **`_data/social_media.yml`** → defines platforms (name, link, icon)

* **`_includes/social_media_share_url.html`** → builds “share post” links

### Purpose:

* Add/remove social platforms
* Control how sharing links are generated

### In short:

One file defines **which social sites exist**, the other defines **how sharing works**.
