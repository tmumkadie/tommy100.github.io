# HOW TO COMMITS

How to commit in VS Code:

1. `Ctrl + Shift + G` → opens Source Control
2. `+` → stage your file
3. Type a message → `"your change description"`
4. Click Commit → then Sync Changes

Done — changes are live in 1-2 minutes! ✅

# Summary

The `/_includes/` folder contains **reusable HTML templates** (like header, footer, cards, and lists) that **take data from `_config.yml` or your posts/projects** and display it dynamically on your site—essentially turning your configuration and content into visible webpage sections.

#----------------------------------------------------------------------------------
#A Liquid file is a template that **combines HTML + placeholders (`{{ }}`) + logic (`{% %}`) with data to produce a rendered page**.

Example:

```liquid
<h1>Hello, {{ name }}!</h1>
```

**Data: `name: Tommy` → Rendered: `<h1>Hello, Tommy!</h1>`
