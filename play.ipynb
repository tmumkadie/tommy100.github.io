# HOW TO COMMITS

How to commit in VS Code:

1. `Ctrl + Shift + G` → opens Source Control
2. `+` → stage your file
3. Type a message → `"your change description"`
4. Click Commit → then Sync Changes

Done — changes are live in 1-2 minutes! ✅


#COMMIT MEANS

Commit = saving your changes to GitHub permanently.
Like hitting "Save" but for your entire project's history.

# GitHub is essentially a cloud storage for your code — every commit is saved there permanently, accessible from anywhere, even if your local computer crashes or files get deleted.
